# 1. Line Plot

```python
# importing matplotlib module 
from matplotlib import pyplot as plt 

# x-axis values 
x = [15, 2, 19, 4, 17] 

# Y-axis values 
y = [10, 5, 8, 4, 2] 

# x label
plt.xlabel("X label")

# y label
plt.ylabel("Y label")

# Function to plot 
plt.plot(x,y,c="black",linestyle='dashed', marker='o',) 

# function to show the plot 
plt.show() 

#Save Figure
plt.savefig("Line Plot.png")
```

# 2. Bar Plot

```python
# x-axis values 
x = [15, 2, 19, 4, 17] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 

# x label
plt.xlabel("X label")

# y label
plt.ylabel("Y label")
  
# Function to plot the bar 
plt.bar(x,y,color="yellow") 
  
# function to show the plot 
plt.show() 

#Save Figure
plt.savefig("Bar Plot.png")
```

# 3. Histogram

```python
# Y-axis values 
y = [20, 5, 8, 4, 2] 
  
# Function to plot histogram 
plt.hist(y,color="green") 

# x label
plt.xlabel("X label")

# y label
plt.ylabel("Y label")

  
# Function to show the plot 
plt.show() 

#Save Figure
plt.savefig("Histogram.png")
```

# 4. Scatter Plot

```python
# importing random  
from numpy import random
  
# x-axis values 
x=random.randint(100, size=(100))
  
# Y-axis values 
y=random.randint(100, size=(100))

# x label
plt.xlabel("X label")

# y label
plt.ylabel("Y label")
  
# Function to plot scatter 
plt.scatter(x, y,c="m") 
  
# function to show the plot 
plt.show() 

#Save Figure
plt.savefig("Scatter Plot.png")
```
